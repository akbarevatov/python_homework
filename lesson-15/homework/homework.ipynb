# Task 2

```python
x2 = np.linspace(0, np.pi, 1000)
y2_1 = np.sin(x2)
y2_2 = np.cos(x2)

plt.plot(x2, y2_1, color="red", linestyle=":", label=r"$ \sin(x) $")
plt.plot(x2, y2_2, color="blue", linestyle="--", label=r"$ \cos(x) $")
plt.legend()
plt.show()
```

# Task 3

```python
x3 = np.linspace(0, 10, 1000)
top_left = x3**3
top_right = np.sin(x3)
bottom_left = np.e**x3
bottom_right = np.where(x3!=0, np.log(x3+1),0)
plt.subplot(2, 2, 1)
plt.plot(x3, top_left, color="black", label=r"$ f(x) = x^3 $")
plt.legend()
plt.subplot(2, 2, 2)
plt.plot(x3, top_right, color="red", label=r"$ f(x) = \sin(x) $")
plt.legend()
plt.subplot(2, 2, 3)
plt.plot(x3, bottom_left, color="blue", label=r"$ f(x) = e^x $")
plt.legend()
plt.subplot(2, 2, 4)
plt.plot(x3, bottom_right, color="green", label=r"$ f(x) = \log(x+1) $ (for $ x \geq 0 $)")
plt.legend()
plt.show()
```

# Task 4

```python
# make a list of random numbers
x4 = np.random.uniform(0, 10, 100)
y4 = np.random.uniform(0, 10, 100)

# Make a list of random markers
markers = ['o', 's', 'D', 'v', '^', '<', '>', 'p', '*', 'h']
marker_choices = np.random.choice(markers, size=100)
plt.figure(figsize=(8, 6))

# Scatter the points
for i in range(100):
    plt.scatter(x4[i], y4[i], marker=marker_choices[i])

# Add Title and axis labels
plt.title("Scatter Plot of 100 Random Points")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
# Add grid
plt.grid(True, linestyle="--", alpha=0.7)
# Show the plot
plt.show()
```
